# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.

```python
df4['customer lifetime value'] = df4['customer lifetime value'].apply(lambda x: float(str(x).rstrip('%'))*100 if '%' in str(x) else x)
```

```python
df4['customer lifetime value'] = df4['customer lifetime value'].astype(float)
```
